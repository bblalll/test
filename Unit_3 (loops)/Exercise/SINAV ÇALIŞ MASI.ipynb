```python
print("*************************\na temiyehoş geldiniz\n*************************")

print("""

işleminizi yapın:

1.bakiye:

2.yatırım:

3.çekme:

geri dönmek için 'q' tuşuna basın

""")

bakiye = 1000

while True:
    işlem = input("işleminizi girin...")
    
    if(işlem == "q"):
        print("yine bekleriz...")
        break

    elif (işlem == "1"):
        print("bakiyeniz {} tl dir".format(bakiye))
        
    elif(işlem == "2"):
        miktar = int(input("ytırımınız para:"))
        bakiye += miktar
        print("bakiyeniz {} tl dir".format(bakiye))    
          
        
    
    
    elif(işlem == "3"):
        miktar = int(input("çektiniz para:"))
        bakiye -= miktar
        print("bakiyeniz {} tl dir".format(bakiye))    
           
         
        
        if(bakiye - miktar < 0):
            print("bu parayı çekemez siniz...")
            print("bakiyeniz {} tl dir".format(bakiye))
        continue
        
        
        bakiye -= miktar
    else:
        print("lütfen geçerliişlem girin.....")
    
```
